```python
# Đọc dữ liệu vào DataFrame
import pandas as pd
df=pd.read_csv(filepath_or_buffer = 'FoodPrice_in_Turkey.csv', sep = ',')
df.head()
```

```python
df.to_excel('demo_FoodPrice.xlsx')
dfxlsx = pd.read_excel('demo_FoodPrice.xlsx')
dfxlsx.head()
```

```python
df.to_csv('demo_FoodPrice.csv')
dftest = pd.read_csv('demo_FoodPrice.csv')
dftest.head()
```

```python
df.to_json('demo_FoodPrice.json',orient='columns')
dfjson = pd.read_json('demo_FoodPrice.json')
dfjson.head()
```

```python
df.to_hdf('demo_FoodPrice.h5', key ='table')
dfhdf = pd.read_hdf('demo_FoodPrice.h5',key= 'table')
dfhdf.head()
```
